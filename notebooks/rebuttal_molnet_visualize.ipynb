```python
import numpy as np
import math
from scipy import stats

def _get_vals_or_nan(input_list):
    rtn = []
    for x in input_list:
        if type(x) == str and '+' in x:
            rtn.append(float(x[0:x.index('+')]))
        elif math.isnan(float(x)):
            rtn.append(0)
    return rtn

def _compute_per_task_auc_pr(vals, frac_positive_test):
    per_task_auc_pr = [(r - f) for (r,f) in zip(vals, frac_positive_test)]
    return sum(per_task_auc_pr) / len(vals), sum(per_task_auc_pr)


def compute_auc(dataset='sider'):
    for support in ['16', '32', '64', '128']:
        if dataset == 'sider':
            s = 5
            e=32
        elif dataset == 'tox21':
            s = 32
            e = 44
        elif dataset == 'bace':
            s= 0
            e = 1
        elif dataset == 'bbbp':
            s = 1
            e = 2
        elif dataset == 'clintox':
            s = 2
            e = 3 # Ending on 4 predicts FDA approval status.
        elif dataset == 'hiv':
            s = 4
            e = 5
        non_nan_idx = np.array(_get_vals_or_nan(data[f'16_train (MT)'][s:e])) != 0
        print(np.count_nonzero(non_nan_idx))
        frac_pos = [float(x) for x in data['fraction_positive_test'][s:e][non_nan_idx].values]
        mt = np.array(_get_vals_or_nan(data[f'{support}_train (MT)'][s:e]))[non_nan_idx]
        mt_dpr = np.mean(mt)
        mt_ste = stats.sem(mt)
        
        ctx = np.array(_get_vals_or_nan(data[f'{support}_train (CONTEXT_{support})'][s:e]))[non_nan_idx]
        ctx_dpr = np.mean(ctx)
        ctx_ste = stats.sem(ctx)
        
        maml = np.array(_get_vals_or_nan(data[f'{support}_train (MAML)'][s:e]))[non_nan_idx]
        maml_dpr = np.mean(maml)
        maml_ste = stats.sem(maml)
        
        proto = np.array(_get_vals_or_nan(data[f'{support}_train (PROTO)'][s:e]))[non_nan_idx]
        proto_dpr = np.mean(proto)
        proto_ste = stats.sem(proto)
        print(f'Support {support}: (MT, {mt_dpr:.3f}+/-{mt_ste:.3f}); (CTX, {ctx_dpr:.3f}+/-{ctx_ste:.3f});'
              f'(MAML, {maml_dpr:.3f}+/-{maml_ste:.3f}); (PROTO, {proto_dpr:.3f}+/-{proto_ste:.3f})')

        
compute_auc(dataset='clintox')

# sider
# idx = np.array(_get_vals_or_nan(data['16_train (MT)'][5:31])) != 0
# print(np.mean(np.array(_get_vals_or_nan(data['16_train (MT)'][5:31]))[idx]))
# print(np.mean(np.array(_get_vals_or_nan(data['16_train (CONTEXT_16)'][5:31]))[idx]))

# idx = np.array(_get_vals_or_nan(data['32_train (MT)'][5:31])) != 0
# print(np.mean(np.array(_get_vals_or_nan(data['32_train (MT)'][5:31]))[idx]))
# print(np.mean(np.array(_get_vals_or_nan(data['32_train (CONTEXT_32)'][5:31]))[idx]))

# tox21
# idx = np.array(_get_vals_or_nan(data['16_train (MT)'][32:43])) != 0
# print(np.mean(np.array(_get_vals_or_nan(data['16_train (MT)'][32:43]))[idx]))
# print(np.mean(np.array(_get_vals_or_nan(data['16_train (CONTEXT_16)'][32:43]))[idx]))
```

```python
import math

def _get_vals(input_list):
    rtn = []
    for x in input_list:
        if type(x) == str and '+' in x:
            rtn.append(float(x[0:x.index('+')]))
    return rtn

def _compute_per_task_auc_pr(vals, frac_positive_test):
    per_task_auc_pr = [(r - f) for (r,f) in zip(vals, frac_positive_test)]
    return sum(per_task_auc_pr) / len(vals), sum(per_task_auc_pr)

def compute_support_results(support_size='16_train'):
    mt_csv = pd.read_csv('baselines/GNN-Multitask_summary.csv')
    reported_vals = _get_vals(mt_csv[support_size].values)
    frac_positive_test = [float(x) for x in mt_csv['fraction_positive_test'].values]
    reported_vals, reported_total = _compute_per_task_auc_pr(reported_vals, frac_positive_test)

    # Load in our computed values.
    mt_vals = _get_vals(data[f'{support_size} (GNN-MT)'].values)
    mt_mean, mt_total = _compute_per_task_auc_pr(mt_vals, frac_positive_test)

    s_emb_mt_vals = _get_vals(data[f'{support_size} (EMB-GNN-MT)'].values)
    s_emb_mt_mean, s_emb_mt_total = _compute_per_task_auc_pr(s_emb_mt_vals, frac_positive_test)

    l_emb_mt_vals = _get_vals(data[f'{support_size} (L-EMB-MT)'].values)
    l_emb_mt_mean, l_emb_mt_total = _compute_per_task_auc_pr(l_emb_mt_vals, frac_positive_test)

    # Print out the mean change.
    print(f'mean delta auc_pr reported: {reported_vals}')
    print(f'mean delta auc_pr mt: {mt_mean}')
    print(f'mean delta auc_pr s-emb-mt: {s_emb_mt_mean}')
    print(f'mean delta auc_pr l-emb-mt: {l_emb_mt_mean}')
    
    # Print out the total change
    print()
    print(f'total delta auc_pr reported: {reported_total - mt_total}')
    print(f'total delta auc_pr mt: {mt_total - mt_total}')
    print(f'total delta auc_pr s-emb-mt: {s_emb_mt_total - mt_total}')
    print(f'total delta auc_pr l-emb-mt: {l_emb_mt_total - mt_total}')
    
    return (reported_vals, mt_mean, s_emb_mt_mean, l_emb_mt_mean)

def compute_proto_support_results(support_size='16_train'):
    mt_csv = pd.read_csv('baselines/ProtoNet-gnn+ecfp+fc-Support64_summary.csv')
    reported_vals = _get_vals(mt_csv[support_size].values)
    frac_positive_test = [float(x) for x in mt_csv['fraction_positive_test'].values]
    reported_vals, reported_total = _compute_per_task_auc_pr(reported_vals, frac_positive_test)

    # Load in our computed values.
    mt_vals = _get_vals(data[f'{support_size} (GNN-PROTO)'].values)
    mt_mean, mt_total = _compute_per_task_auc_pr(mt_vals, frac_positive_test)

    s_emb_mt_vals = _get_vals(data[f'{support_size} (GNN-PROTO)'].values)
    s_emb_mt_mean, s_emb_mt_total = _compute_per_task_auc_pr(s_emb_mt_vals, frac_positive_test)

    l_emb_mt_vals = _get_vals(data[f'{support_size} (L-EMB-GNN-PROTO)'].values)
    l_emb_mt_mean, l_emb_mt_total = _compute_per_task_auc_pr(l_emb_mt_vals, frac_positive_test)

    # Print out the mean change.
    print(f'mean delta auc_pr reported: {reported_vals}')
    print(f'mean delta auc_pr mt: {mt_mean}')
    print(f'mean delta auc_pr s-emb-mt: {s_emb_mt_mean}')
    print(f'mean delta auc_pr l-emb-mt: {l_emb_mt_mean}')
    
    # Print out the total change
    print()
    print(f'total delta auc_pr reported: {reported_total - mt_total}')
    print(f'total delta auc_pr mt: {mt_total - mt_total}')
    print(f'total delta auc_pr s-emb-mt: {s_emb_mt_total - mt_total}')
    print(f'total delta auc_pr l-emb-mt: {l_emb_mt_total - mt_total}')
    
    return (reported_vals, mt_mean, s_emb_mt_mean, l_emb_mt_mean)

def compute_maml_support_results(support_size='16_train'):
    mt_csv = pd.read_csv('baselines/MAML-Support16_summary.csv')
    reported_vals = _get_vals(mt_csv[support_size].values)
    frac_positive_test = [float(x) for x in mt_csv['fraction_positive_test'].values]
    reported_vals, reported_total = _compute_per_task_auc_pr(reported_vals, frac_positive_test)

    # Load in our computed values.
    mt_vals = _get_vals(data[f'{support_size} (L-EMB-MAML-2)'].values)
    mt_mean, mt_total = _compute_per_task_auc_pr(mt_vals, frac_positive_test)

#     s_emb_mt_vals = _get_vals(data[f'{support_size} (MAML)'].values)
#     s_emb_mt_mean, s_emb_mt_total = _compute_per_task_auc_pr(s_emb_mt_vals, frac_positive_test)

    l_emb_mt_vals = _get_vals(data[f'{support_size} (L-EMB-MAML-3)'].values)
    l_emb_mt_mean, l_emb_mt_total = _compute_per_task_auc_pr(l_emb_mt_vals, frac_positive_test)

    # Print out the mean change.
    print(f'mean delta auc_pr reported: {reported_vals}')
    print(f'mean delta auc_pr maml: {mt_mean}')
    print(f'mean delta auc_pr l-emb-mt: {l_emb_mt_mean}')
    
    # Print out the total change
    print()
    print(f'total delta auc_pr reported: {reported_total - mt_total}')
    print(f'total delta auc_pr mt: {mt_total - mt_total}')
    print(f'total delta auc_pr l-emb-mt: {l_emb_mt_total - mt_total}')
    
    return (reported_vals, mt_mean, l_emb_mt_mean)

def compute_maml_support_results_v2(support_size='16_train'):
    mt_csv = pd.read_csv('baselines/MAML-Support16_summary.csv')
    reported_vals = _get_vals(mt_csv[support_size].values)
    frac_positive_test = [float(x) for x in mt_csv['fraction_positive_test'].values]
    reported_vals, reported_total = _compute_per_task_auc_pr(reported_vals, frac_positive_test)

    # Load in our computed values.
    l_emb_maml_2 = _get_vals(data[f'{support_size} (L-EMB-MAML-2)'].values)
    l_emb_maml_2_mean, _ = _compute_per_task_auc_pr(l_emb_maml_2, frac_positive_test)

#     s_emb_mt_vals = _get_vals(data[f'{support_size} (MAML)'].values)
#     s_emb_mt_mean, s_emb_mt_total = _compute_per_task_auc_pr(s_emb_mt_vals, frac_positive_test)

    l_emb_maml_3 = _get_vals(data[f'{support_size} (L-EMB-MAML-3)'].values)
    l_emb_maml_3_mean, _ = _compute_per_task_auc_pr(l_emb_maml_3, frac_positive_test)

    # Print out the mean change.
    print(f'mean delta auc_pr l_emb_maml_2: {l_emb_maml_2_mean}')
    print(f'mean delta auc_pr l_emb_maml_3: {l_emb_maml_3_mean}')
    print(f'mean delta auc_pr [reported]: {reported_vals}')
    print(f'differenc in delta auc_pr [l_emb_maml_3 - reported]: {l_emb_maml_3_mean - reported_vals}')
    
    # Print out the total change
#     print()
#     print(f'total delta auc_pr reported: {reported_total - mt_total}')
#     print(f'total delta auc_pr mt: {mt_total - mt_total}')
#     print(f'total delta auc_pr l-emb-mt: {l_emb_mt_total - mt_total}')
    return None
#     return (reported_vals, mt_mean, l_emb_mt_mean)
```
